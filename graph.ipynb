# DFS e BFS

```python
from collections import defaultdict 
  
class Graph: 
  
    def __init__(self): 
  
        self.graph = defaultdict(list) 
  
    def addEdge(self, u, v): 
        self.graph[u].append(v) 
  
    def DFSUtil(self, v, visited): 
  
        visited[v] = True
        print(v, end = ' ') 

        for i in self.graph[v]: 
            if visited[i] == False: 
                self.DFSUtil(i, visited) 

    def DFS(self, v): 

        visited = [False] * (len(self.graph)) 
  
        self.DFSUtil(v, visited) 

    def BFS(self, s): 
  
        visited = [False] * (len(self.graph)) 
  
        queue = [] 

        queue.append(s) 
        visited[s] = True
  
        while queue: 

            s = queue.pop(0) 
            print (s, end = " ") 

            for i in self.graph[s]: 
                if visited[i] == False: 
                    queue.append(i) 
                    visited[i] = True
```

## Visualization of the generated Graph

```python
G = nx.Graph()
g = Graph() 

n = 50
for i in range(n):
    a = randint(0, 20)
    b = randint(0, 20)
    G.add_edge(a, b)
    g.addEdge(a, b)

nx.draw(G, with_labels=True)
plt.show()
```

# Cross

```python
print("Following is DFS from (starting from vertex {})".format(a)) 
g.DFS(a) 
print('\n')
print("Following is BFS from (starting from vertex {})".format(a)) 
g.BFS(a)
```

```python
# Isolate the sequence of degrees
degree_sequence = list(G.degree())
degree_sequence
```

```python
nb_nodes = n
nb_arr = len(G.edges())
avg_degree = np.mean(np.array(degree_sequence)[:,1])
med_degree = np.median(np.array(degree_sequence)[:,1])
max_degree = max(np.array(degree_sequence)[:,1])
min_degree = np.min(np.array(degree_sequence)[:,1])
```

```python
print("Number of nodes : " + str(nb_nodes))
print("Number of edges : " + str(nb_arr))
print("Maximum degree : " + str(max_degree))
print("Minimum degree : " + str(min_degree))
print("Average degree : " + str(avg_degree))
print("Median degree : " + str(med_degree))
```
